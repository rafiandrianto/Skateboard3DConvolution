```python
import os
import subprocess
```

```python
def convert_and_resize(source_dir, target_dir=None, resolution=(640, 480)):
    if target_dir is None:
        target_dir = source_dir

    if not os.path.exists(target_dir):
        os.makedirs(target_dir)

    files = [f for f in os.listdir(source_dir) if f.endswith('.mov')]
    if not files:
        print("No .mov files found in the directory.")
        return

    for file in files:
        full_path = os.path.join(source_dir, file)
        target_path = os.path.join(target_dir, f"{os.path.splitext(file)[0]}.mp4")
        # includes resolution change
        command = ["ffmpeg", "-i", full_path, "-vf", f"scale={resolution[0]}:{resolution[1]}", "-q:v", "0", target_path]

        try:
            subprocess.run(command, check=True, stdout=subprocess.PIPE, stderr=subprocess.PIPE)
            print(f"Successfully converted and resized {file} to {os.path.splitext(file)[0]}.mp4")
        except subprocess.CalledProcessError as e:
            print(f"Failed to convert and resize {file}. Error: {e.stderr.decode()}")
```

```python
source = '/home/rafi/Desktop/GradResearch/ZeroSkate/Tricks/Kickflip'
new = '/home/rafi/Desktop/GradResearch/ZeroSkate/NewTricks/Kickflip'
convert_and_resize(source, new)
```
